```python
# (some word with prefix/suffix) --> Lemmetisation --> Lemma (root word)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmetiser = WordNetLemmatizer()
```

```python
'''
POS parameter:
Noun - n
Verb - v
Adjective - a
Adverb - r

'''


lemmetiser.lemmatize('history')
```

```python
lemmetiser.lemmatize('historical', pos='a')
```

```python
lemmetiser.lemmatize('historically', pos = 'r')
```

```python

words = ["eating","eaten","eats","writing","writes","programming","programs","programmers","story"]
```

```python
for word in words:
    print(word+" --> "+lemmetiser.lemmatize(word, pos='v'))
```
