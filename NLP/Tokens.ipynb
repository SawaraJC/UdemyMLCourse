```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
corpus ="""Traditionally, the authorship of the Mahābhārata is attributed to Vyāsa. There have been many attempts to unravel its historical growth and compositional layers. The bulk of the Mahābhārata was probably compiled between the 3rd century BCE and the 3rd century CE, with the oldest preserved parts not much older than around 400 BCE.[6][7] The text probably reached its final form by the early Gupta period (c. 4th century CE)"""
```

```python
print(corpus)
```

```python
##tokenisation: sentence to paragraph

from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize(corpus)
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentences in documents:
    print(sentences)
```

```python
## paragragh --> words
## sentence --> words

from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentences in documents:
    print(word_tokenize(sentences))
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
