```python
# stem: root word
words = ["eating","eaten","eats","writing","writes","programming","programs","programmers","story"]
```

## porter stemming

```python

from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+" --> "+stemming.stem(word))
    
```

## regex stemmer

```python
from nltk.stem import RegexpStemmer
```

```python
reg = RegexpStemmer('ing$|s$|able$', min=4)
```

```python
reg.stem('eating')
```

```python
reg.stem('eats')
```

```python
reg.stem('eatable')
```

## snowball stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snow = SnowballStemmer('english')
```

```python
for word in words:
    print(word+" --> "+snow.stem(word))
```

```python
stemming.stem('fairly')
```

```python
snow.stem('fairly')
```
