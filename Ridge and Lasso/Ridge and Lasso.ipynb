## Change required columns as integer data type


```python
df1[['month','day','year','Temperature','RH','Ws']] = df1[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df1.iloc[[122]] #not cleared, bcz we used dropna
```

```python
df1 = df1.drop(index=122).reset_index(drop=True)
```

```python
df1.iloc[[122]] #now done
```

```python
df1[['month','day','year','Temperature','RH','Ws']] = df1[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df1.info()
```

## Changing remaining columns to float

```python
objects = [features for features in df1.columns if df1[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df1[i] = df1[i].astype(float) #skipped classes, since it is a categorical feature
```
